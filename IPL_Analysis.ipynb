```python
# Create a mapping dictionary for old team names to new names
team_name_mapping = {
    'Kings XI Punjab': 'Punjab Kings',
    'Delhi Daredevils': 'Delhi Capitals',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Royal Challengers Bengaluru': 'Royal Challengers Bangalore'
    # Add more mappings if needed
}

# Replace old team names with new names in 'team1', 'team2', and 'winner' columns
matches['team1'] = matches['team1'].replace(team_name_mapping)
matches['team2'] = matches['team2'].replace(team_name_mapping)
matches['winner'] = matches['winner'].replace(team_name_mapping)

# Verify the changes
print(matches[['team1', 'team2', 'winner']].head())
```

```python
#lets combine this 
combined_team  = pd.concat( [matches['team1'],matches['team2']])
print(combined_team)
print(" ")
matches_by_each_team = combined_team.value_counts()
print(matches_by_each_team)
```

```python
balls_bowled = balls.groupby('bowler')['ball'].count().sort_values(ascending=False)
print(balls_bowled)
```

```python
runs_given = balls.groupby('bowler')['total_runs'].sum().sort_values(ascending=False)
print(runs_given)
```

```python
wickets_taken = balls.groupby('bowler')['is_wicket'].sum().sort_values(ascending = False)
print(wickets_taken)
```

```python
most_runs = balls.groupby('batter')['batsman_runs'].sum().sort_values(ascending = False)
print(most_runs)
```

```python
balls_played = balls.groupby('batter')['ball'].count().sort_values(ascending=False)
print(balls_played)
```

```python
# Filter batters who have played more than 50 balls
filtered_batters = balls_played[balls_played > 50]

# Calculate strike rate only for these batters
strike_rate = (most_runs[filtered_batters.index] / filtered_batters) * 100

# Sort the strike rate in descending order
strike_rate = strike_rate.sort_values(ascending=False)

print(strike_rate)
```

```python
career_stats = pd.DataFrame({
    'Total Runs': most_runs[filtered_batters.index],
    'Balls Faced': filtered_batters,
    'Strike Rate': strike_rate
}).sort_values(by='Total Runs', ascending=False)

print(career_stats)
```
